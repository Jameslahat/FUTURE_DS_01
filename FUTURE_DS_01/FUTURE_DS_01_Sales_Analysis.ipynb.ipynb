```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

monthly_sales.plot(figsize=(10,5), title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

**Sales show an upward trend during year-end months, indicating seasonal demand.**

```python
top_products = (
    df.groupby('Product Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products.plot(kind='barh', figsize=(8,6), title='Top 10 Products by Sales')
plt.xlabel('Sales')
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Sales by Category')
plt.ylabel('')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()

region_sales.plot(kind='bar', figsize=(6,4), title='Sales by Region')
plt.ylabel('Sales')
plt.show()
```

```python
top_states = (
    df.groupby('State')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_states.plot(kind='bar', figsize=(8,4), title='Top 10 States by Sales')
plt.show()
```

**The West region generates the highest sales, indicating strong customer demand.**

**Technology and Furniture categories contribute the most to overall revenue.**

**Sales peak during year-end months, suggesting seasonal buying behavior.**

**A small number of products generate a disproportionate share of total revenue, following the Pareto principle.**

**Certain states significantly outperform others, indicating regional market concentration.**
